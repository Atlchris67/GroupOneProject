```python
base_url = "https://api.exchangeratesapi.io/history?"
params = {
    "start_at": "2019-01-01",  
    "end_at": "2020-01-02",
    "symbols": "USD,CNY"
}
response = requests.get(base_url, params=params).json()
```

```python
rates_dt = response['rates']
#print(rates_dt)
#type(response['rates'])
#print(rates_dt)

cols = ['Date', "USD", ]
df2 = pd.DataFrame(columns=cols)
for key, value in rates_dt.items():
    df2 = df2.append({'Date': key, 'USD': value.get("USD"), 'CNY': value.get("CNY")}, ignore_index=True)   
    
df2 = df2.sort_values("Date")    
df2.head()


df2.to_csv("exchangeRates.csv")
```
